```python
import numpy as np

def calculate_distance(point1, point2):

  distance = np.sqrt(np.sum((point1 - point2)**2))
  return distance

# Combine dataX and dataY into a 2D array of points
data_points = np.column_stack((dataX, dataY))

# Calculate distances between all pairs of points
num_points = data_points.shape[0]
distances = np.zeros((num_points, num_points))

for i in range(num_points):
  for j in range(i + 1, num_points):  # Avoid redundant calculations
    distances[i, j] = calculate_distance(data_points[i], data_points[j])
    distances[j, i] = distances[i, j]  # Distance is symmetric

# Print the distance matrix
print("Distance Matrix:")
print(distances)
```
